```python
def regions_from_graph(nodes, edges):
    """
    Extract regions from nodes and edges of a planar graph

    Arguments
    ---------

    nodes: dictionary with vertex id as key, coordinates of vertex as value

    edges: list of (head,tail), (tail, head) edges

    Returns
    ------

    regions: list of lists of nodes defining a region. Includes the external
    region
    filaments: list of lists of nodes defining filaments and isolated
    vertices



    Examples
    --------
    >>> vertices = {0: (1, 8), 1: (1, 7), 2: (4, 7), 3: (0, 4), 4: (5, 4), 5: (3, 5), 6: (2, 4.5), 7: (6.5, 9), 8: (6.2, 5), 9: (5.5, 3), 10: (7, 3), 11: (7.5, 7.25), 12: (8, 4), 13: (11.5, 7.25), 14: (9, 1), 15: (11, 3), 16: (12, 2), 17: (12, 5), 18: (13.5, 6), 19: (14, 7.25), 20: (16, 4), 21: (18, 8.5), 22: (16, 1), 23: (21, 1), 24: (21, 4), 25: (18, 3.5), 26: (17, 2), 27: (19, 2)}
    >>> edges = [(1, 2),(1, 3),(2, 1),(2, 4),(2, 7),(3, 1),(3, 4),(4, 2),(4, 3),(4, 5),(5, 4),(5, 6),(6, 5),(7, 2),(7, 11),(8, 9),(8, 10),(9, 8),(9, 10),(10, 8),(10, 9),(11, 7),(11, 12),(11, 13),(12, 11),(12, 13),(12, 20),(13, 11),(13, 12),(13, 18),(14, 15),(15, 14),(15, 16),(16, 15),(18, 13),(18, 19),(19, 18),(19, 20),(19, 21),(20, 12),(20, 19),(20, 21),(20, 22),(20, 24),(21, 19),(21, 20),(22, 20),(22, 23),(23, 22),(23, 24),(24, 20),(24, 23),(25, 26),(25, 27),(26, 25),(26, 27),(27, 25),(27, 26)]
    >>> r = regions_from_graph(vertices, edges)
    >>> r['filaments']
    [[6, 5, 4], 0, [2, 7, 11], [14, 15, 16], 17]
    >>> r['regions']
    [[3, 4, 2, 1, 3], [9, 10, 8, 9], [11, 12, 13, 11], [12, 20, 19, 18, 13, 12], [19, 20, 21, 19], [22, 23, 24, 20, 22], [26, 27, 25, 26]]

    Notes
    -----
    Based on
    Eberly http://www.geometrictools.com/Documentation/MinimalCycleBasis.pdf.
    """
    def find_start_node(nodes,node_coord):
        start_node = []
        minx = float('inf')
        for key,node in nodes.items():
            if node[0] <= minx:
                minx = node[0]
                start_node.append(key)
        if len(start_node) > 1:
            miny = float('inf')
            for i in range(len(start_node)):
                if nodes[i][1] < miny:
                    miny = nodes[i][1]
                else:
                    start_node.remove(i)
        return nodes[start_node[0]], node_coord[nodes[start_node[0]]]
    
    def adj_nodes(start_key, edges):
        start_key
        vnext = []
        for edge in edges:
            if edge[0] == start_key:
                vnext.append(edge[1])
        if len(vnext) == 0:
            pass
            #print "Vertex is end point."
        return vnext
    
    def remove_edge(v0,v1,edges, ext_edges, node2link, startnode, end_node, filament=False):
        #Start to build the WED
        if filament == True:
            if v0 == v1: #This handles the case of node 16, where it is isolated and edgeless after eberly.
                for edge in ext_edges:
                    if edge[0] == v0:
                        v1 = edge[1]
                        node2link[v0] = (v0,v1)
                        startnode[(v0,v1)] = v0
                        end_node[(v0,v1)] = v1
                        break
            else:
                node2link[v0] = (v0,v1)
                node2link[v1] = (v1,v0)
                startnode[(v0,v1)] = v0
                end_node[(v0,v1)] = v1
        
        #Eberly
        try:
            ext_edges.append((v0,v1))
            ext_edges.append((v1,v0))
            edges.remove((v0,v1))
            edges.remove((v1,v0))
        except:
            pass
        return edges, ext_edges,node2link, startnode, end_node
    
    def remove_heap(v0,sorted_nodes):
        sorted_nodes[:] = [x for x in sorted_nodes if x[0] != v0]
        return sorted_nodes
            
    def remove_node(v0, nodes, nodes_coord, vertices):
        vertices[v0] = nodes[v0]
        del nodes_coord[nodes[v0]]
        del nodes[v0]
        return nodes, nodes_coord, vertices
    
    def clockwise(nodes,vnext,start_key, v_prev):
        v_curr = np.asarray(nodes[start_key])
        v_next = None
        if v_prev == None:
            v_prev = np.asarray([0,-1]) #This should be a vertical tangent to the start node at initialization.
        else:
            pass
        d_curr = v_curr - v_prev
        
        for v_adj in vnext:
            #No backtracking
            if np.array_equal(np.asarray(nodes[v_adj]),v_prev) == True:
                continue
            
            #The potential direction to move in
            d_adj = np.asarray(nodes[v_adj]) - v_curr
            
            #Select the first candidate
            if v_next is None:
                v_next = np.asarray(nodes[v_adj])
                d_next = d_adj
                convex = d_next[0]*d_curr[1] - d_next[1]*d_curr[0]
                if convex <= 0:
                    convex = True
                else:
                    convex = False
            
            #Update if the next candidate is clockwise of the current clock-wise most
            if convex == True:
                if (d_curr[0]*d_adj[1] - d_curr[1]*d_adj[0]) < 0 or (d_next[0]*d_adj[1]-d_next[1]*d_adj[0]) < 0:
                    v_next = np.asarray(nodes[v_adj])
                    d_next = d_adj 
                    convex = d_next[0]*d_curr[1] - d_next[1]*d_curr[0]
                    if convex <= 0:
                        convex = True
                    else:
                        convex = False
            else:
                if (d_curr[0]*d_adj[1] - d_curr[1]*d_adj[0]) < 0 and (d_next[0]*d_adj[1]-d_next[1]*d_adj[0]) < 0:
                    v_next = np.asarray(nodes[v_adj])
                    d_next = d_adj 
                    convex = d_next[0]*d_curr[1] - d_next[1]*d_curr[0] 
                    if convex <= 0:
                        convex = True
                    else:
                        convex = False
        prev_key = start_key
        return tuple(v_next.tolist()), node_coord[tuple(v_next.tolist())], prev_key
    
    def counterclockwise(nodes, vnexts, start_key, prev_key):
        v_next = None
        v_prev = np.asarray(nodes[prev_key])
        v_curr = np.asarray(nodes[start_key])
        d_curr = v_curr - v_prev
        
        for v_adj in vnexts: 
            #Prohibit Back-tracking
            if v_adj == prev_key:
                continue
            d_adj = np.asarray(nodes[v_adj]) - v_curr
            
            if v_next == None:
                v_next = np.asarray(nodes[v_adj])
                d_next = d_adj
                convex = d_next[0]*d_curr[1] - d_next[1]*d_curr[0]
                
            if convex <= 0:
                if d_curr[0]*d_adj[1] - d_curr[1]*d_adj[0] > 0 and d_next[0]*d_adj[1] - d_next[1]*d_adj[0] > 0:
                    v_next = np.asarray(nodes[v_adj])
                    d_next = d_adj 
                    convex = d_next[0]*d_curr[1] - d_next[1]*d_curr[0]
                else:
                    pass
            else:
                if d_curr[0]*d_adj[1] - d_curr[1]*d_adj[0] > 0 or d_next[0]*d_adj[1]-d_next[1]*d_adj[0] > 0:
                    v_next = np.asarray(nodes[v_adj])
                    d_next = d_adj 
                    convex = d_next[0]*d_curr[1] - d_next[1]*d_curr[0]
                else:
                    pass
        prev_key = start_key
        try:
            return tuple(v_next.tolist()), node_coord[tuple(v_next.tolist())], prev_key
        except: 
            return v_next, None, prev_key
        
    def extractisolated(nodes,node_coord,v0,primitives, vertices, ext_edges):
        #primitives.append(v0)
        nodes, node_coord, vertices = remove_node(v0, nodes, node_coord, vertices)
        return nodes, node_coord, primitives, vertices, ext_edges
    
    def extractfilament(v0,v1, nodes, node_coord,sorted_nodes, edges, primitives,cycle_edge, vertices, ext_edges, node2link, startnode, end_node, iscycle=False):
        if (v0,v1) in cycle_edge or (v1,v0) in cycle_edge:
            iscycle = True
        if iscycle == True:
        #This deletes edges that are part of a cycle, but does not add them as primitives.
            if len(adj_nodes(v0,edges)) >= 3:
                edges, ext_edges,node2link, startnode, end_node = remove_edge(v0,v1,edges, ext_edges, node2link, startnode, end_node)
                v0 = v1
                if len(adj_nodes(v0, edges)) == 1:
                    v1 = adj_nodes(v0, edges)[0]
            while len(adj_nodes(v0, edges)) == 1:
                v1 = adj_nodes(v0, edges)[0]
                #Here I need to do the cycle check again.
                iscycle = False
                if (v0,v1) in cycle_edge or (v1,v0) in cycle_edge:
                    iscycle = True
                if iscycle == True:
                    edges, ext_edges,node2link, startnode, end_node = remove_edge(v0,v1,edges, ext_edges, node2link, startnode, end_node)
                    nodes, node_coord, vertices = remove_node(v0, nodes, node_coord, vertices)
                    sorted_nodes = remove_heap(v0, sorted_nodes)
                    v0 = v1
                else:
                    break
            if len(adj_nodes(v0, edges)) == 0:
                nodes, node_coord, vertices = remove_node(v0, nodes, node_coord, vertices)
                sorted_nodes = remove_heap(v0, sorted_nodes)
        else:
            #Filament found
            primitive = []
            if len(adj_nodes(v0,edges)) >= 3:
                primitive.append(v0)
                edges, ext_edges,node2link, startnode, end_node = remove_edge(v0,v1,edges, ext_edges, node2link, startnode, end_node, filament=True)
                v0 = v1
                if len(adj_nodes(v0, edges)) == 1:
                    v1 = adj_nodes(v0, edges)[0]
    
            while len(adj_nodes(v0, edges)) == 1:
                primitive.append(v0)
                v1 = adj_nodes(v0, edges)[0]
                sorted_nodes = remove_heap(v0, sorted_nodes)
                edges, ext_edges,node2link, startnode, end_node = remove_edge(v0, v1, edges, ext_edges, node2link, startnode, end_node, filament=True)
                nodes, node_coord, vertices = remove_node(v0, nodes, node_coord, vertices)
                v0 = v1
            
            primitive.append(v0)
            if len(adj_nodes(v0, edges)) == 0:
                sorted_nodes = remove_heap(v0, sorted_nodes)
                edges, ext_edges,node2link, startnode, end_node = remove_edge(v0, v1, edges, ext_edges, node2link, startnode, end_node, filament=True)
                nodes, node_coord, vertices = remove_node(v0, nodes, node_coord, vertices)
            primitives.append((primitive))
        return sorted_nodes, edges, nodes, node_coord, primitives, vertices, ext_edges,node2link, startnode, end_node
    
    def extract_primitives(start_key,sorted_nodes, edges, nodes, node_coord, primitives,minimal_cycles,cycle_edge, vertices, ext_edges,node2link, region2link, startnode, end_node,left_region):
        v0 = start_key
        visited = []
        sequence = []
        sequence.append(v0)
    
        #Find the CWise most vertex
        vnext = adj_nodes(start_key, edges)
        start_node,v1,v_prev = clockwise(nodes,vnext,start_key,prev_key)
        v_curr = v1
        v_prev = v0
        #Find minimal cycle using CCWise rule
        process = True
        if v_curr == None:
            process = False
        elif v_curr == v0:
            process = False
        elif v_curr in visited:
            process = False
        
        while process == True:
            sequence.append(v_curr)
            visited.append(v_curr)
            vnext = adj_nodes(v_curr, edges)
            v_curr_coords,v_next,v_prev = counterclockwise(nodes,vnext,v_curr, v_prev)
            v_curr = v_next
            if v_curr == None:
                process = False
            elif v_curr == v0:
                process = False
            elif v_curr in visited:
                process = False
        
        if v_curr is None:
            #Filament found, not necessarily at start_key
            sorted_nodes, edges, nodes, node_coord, primitives, vertices, ext_edges,node2link, startnode, end_node = extractfilament(v_prev, adj_nodes(v_prev, edges)[0],nodes, node_coord, sorted_nodes, edges, primitives, cycle_edge, vertices, ext_edges, node2link, startnode, end_node)
            
        elif v_curr == v0:
            #Minimal cycle found
            primitive = []
            iscycle=True
            sequence.append(v0)
            minimal_cycles.append(list(sequence))
            #Remove the v0, v1 edges from the graph.
            edges, ext_edges,node2link, startnode, end_node = remove_edge(v0,v1,edges, ext_edges,node2link, startnode, end_node)
            sorted_nodes = remove_heap(v0, sorted_nodes)#Not in pseudo-code, but in source.
            #Mark all the edges as being part of a minimal cycle.
            if len(adj_nodes(v0, edges)) == 1:
                cycle_edge.append((v0, adj_nodes(v0, edges)[0]))
                cycle_edge.append((adj_nodes(v0, edges)[0], v0))
                sorted_nodes, edges, nodes, node_coord, primitives, vertices, ext_edges,node2link, startnode, end_node = extractfilament(v0, adj_nodes(v0, edges)[0],nodes, node_coord, sorted_nodes, edges, primitives,cycle_edge, vertices, ext_edges,node2link, startnode, end_node)
            if len(adj_nodes(v1, edges)) == 1:
                cycle_edge.append((v1, adj_nodes(v1, edges)[0]))
                cycle_edge.append((adj_nodes(v1, edges)[0],v1))
                sorted_nodes, edges, nodes, node_coord, primitives, vertices, ext_edges,node2link, startnode, end_node = extractfilament(v1, adj_nodes(v1, edges)[0],nodes, node_coord, sorted_nodes, edges, primitives, cycle_edge, vertices, ext_edges,node2link, startnode, end_node)
                
            for i,v in enumerate(sequence[1:-1]):
               cycle_edge.append((v,sequence[i]))
                #cycle_edge.append((sequence[i],v))
            
            #WED
            region2link[len(minimal_cycles)-1] = (minimal_cycles[-1][0], minimal_cycles[-1][1])
            for i in range(len(minimal_cycles[-1])-1):
                left_region[(minimal_cycles[-1][i], minimal_cycles[-1][i+1])] = len(minimal_cycles)-1
            
        else:
            #vcurr was visited earlier, so traverse the filament to find the end
            while len(adj_nodes(v0,edges)) == 2:
                if adj_nodes(v0,edges)[0] != v1:
                    v1 = v0
                    v0 = adj_nodes(v0,edges)[0]
                else:
                    v1 = v0
                    v0 = adj_nodes(v0, edges)[1]
            sorted_nodes, edges, nodes, node_coord, primitives,node2link, startnode, end_node = extractfilament(v0,v1,nodes, node_coord, sorted_nodes, edges, primitives,cycle_edge,node2link, startnode, end_node)
        return sorted_nodes, edges, nodes, node_coord, primitives, minimal_cycles,cycle_edge, vertices, ext_edges, node2link,region2link, startnode, end_node,left_region
    
    wed = {}
    node2link = {}
    region2link = {}
    startnode = {}
    end_node = {}
    left_region = {}
    right_region = {}
    
    #1.
    sorted_nodes = sorted(nodes.iteritems(), key=operator.itemgetter(1))
    node_coord = dict (zip(nodes.values(),nodes.keys()))
    
    #2.
    primitives = []
    minimal_cycles = []
    cycle_edge = []
    prev_key = None #This is only true for the first iteration.
    #This handles edge and node deletion we need populated later.
    vertices = {}
    ext_edges = []    
    
    #3.
    while sorted_nodes: #Iterate through the sorted list
        start_key = sorted_nodes[0][0]
        numadj = len(adj_nodes(start_key, edges))
        if numadj == 0:
            nodes, node_coord, primitives, vertices, ext_edges = extractisolated(nodes,node_coord,start_key,primitives, vertices, ext_edges)
            sorted_nodes.pop(0)
        elif numadj == 1:
            sorted_nodes, edges, nodes, node_coord, primitives, vertices, ext_edges,node2link, startnode, end_node = extractfilament(start_key, adj_nodes(start_key, edges)[0],nodes, node_coord, sorted_nodes, edges,primitives,cycle_edge, vertices, ext_edges,node2link, startnode, end_node)
        else:
            sorted_nodes, edges, nodes, node_coord, primitives, minimal_cycles,cycle_edge, vertices, ext_edges,node2link, region2link, startnode, end_node,left_region = extract_primitives(start_key,sorted_nodes, edges, nodes, node_coord, primitives, minimal_cycles,cycle_edge, vertices, ext_edges, node2link, region2link, startnode, end_node,left_region)
                
    '''By here the entire graph is extracted.  
       All additional WED info is derived from minimal cycles and filaments.'''
    ####START NODES, END NODES, & NODE2LINK FOR ALL LINKS IN MIN CYCLES####
    checked = []
    for cycle in minimal_cycles:
        for i in range(len(cycle)-1):
            if (cycle[i],cycle[i+1]) not in checked and  (cycle[i+1],cycle[i]) not in checked:
                startnode[(cycle[i], cycle[i+1])] = cycle[i]
                end_node[(cycle[i], cycle[i+1])] = cycle[i+1]
                node2link[cycle[i]] = (cycle[i], cycle[i+1])
                checked.append((cycle[i],cycle[i+1]))
                checked.append((cycle[i+1],cycle[i]))
    
    ####FILAMENT REGIONS####
    # build polygon spatial index
    polys = []
    for cycle in minimal_cycles:
        polys.append(ps.cg.Polygon([ps.cg.Point(vertices[pnt]) for pnt in cycle]))
    
    #Find the remaining left_regions once the full graph is extracted.
    for filament in primitives:
        region_flag = False
        end = ps.cg.Point(vertices[filament[-1]])
        for i, poly in enumerate(polys):
            if poly.contains_point(end):
                for j in range(len(filament)-1):
                    right_region[(filament[j], filament[j+1])] = i
                    left_region[(filament[j], filament[j+1])] = i
                region_flag = True
                break         
        if region_flag == False:  
            for j in range(len(filament)-1):
                right_region[(filament[j], filament[j+1])] = -1
                left_region[(filament[j], filament[j+1])] = -1
    
    ####HOLES####
    #Serge's code direct from faces.
    pl = ps.cg.PolygonLocator(polys)
    
    # find all overlapping polygon mbrs
    overlaps ={}
    nump = len(minimal_cycles)
    for i in range(nump):
        overlaps[i] = pl.overlapping(polys[i].bounding_box)
    
    # for overlapping mbrs (left,right) check if right polygon is contained in left
    holes = []
    for k in overlaps:
        for  pc in overlaps[k]:
            s = sum( [polys[k].contains_point(v) for v in pc.vertices])
            if s == len(pc.vertices):
                # print k, pc
                holes.append((k,pc))
                
    # for hole edges add right polygon
    h = {}
    for hole in holes:
        outer, inner = hole
        inner = polys.index(inner)
        #print cycles[outer], cycles[inner]
        if outer in h:
            h[outer].append(inner)
        else:
            h[outer] = [inner]
        nv = len(minimal_cycles[inner])-1
        for i in range(nv):
            edge = minimal_cycles[inner][i],minimal_cycles[inner][i+1]
            right_region[edge] = outer

    ####ADJACENT EDGES####
    start_keys = [k for k in startnode.keys()]
    left_keys = [k for k in left_region.keys()]
    adj_cycles = []
    for edge in cycle_edge:
        if edge in start_keys and (edge[1], edge[0]) in left_keys:
            adj_cycles.append(edge)
            adj_cycles.append((edge[1], edge[0]))
    for link in adj_cycles:
        if link in start_keys:
            pass
        else:
            right_region[(link[1], link[0])] = left_region[link]
            del left_region[link]
    
    ####External EDGES####
    left_keys = [k for k in left_region.keys()]
    right_keys = [k for k in right_region.keys()]
    external_edges = set(left_keys) - set(right_keys)
    for edge in external_edges:
        right_region[edge] = -1 #The external poly.
    
    ####START C and CC LINKS FOR CYCLES####
    start_c_link = {}
    start_cc_link = {}
    for key, value in startnode.iteritems():
        incident_links = []
        for edge in start_keys:
            if value == edge[0] or value == edge[1]:
                if edge != key:
                    incident_links.append(edge)
        if len(incident_links) == 1:
            start_c_link[key] = incident_links[0]
            start_cc_link[key] = incident_links[0]
        else:
            nodes2check = adj_nodes(value, ext_edges)
            nodes2check[:] = [x for x in nodes2check if x != key[1]]
            lr = left_region[key]
            rr = right_region[key]
            for n in nodes2check:
                #Set the CC Link
                try:
                    if left_region[(n, value)] == lr:
                        if left_region[(n,value)] == right_region[(n,value)]:
                            pass
                        else:
                            start_cc_link[key] = (n,value)
                            #print key, start_cc_link[key]
                except:
                    pass
                
                #Set the C Link - NOT WORKING YET
                print right_region[key], right_region[(n,value)]
                try:
                    if right_region[(value,n)] == rr:
                        pass
                        #print key, value,n
                except:
                    pass
                
            
            
    ####END C and CC LINKS FOR CYCLES####
    end_c_link = {}
    end_cc_link = {}
    for key, value in end_node.iteritems():
        incident_links = []
        for edge in start_keys:
            if value == edge[0] or value == edge[1]:
                if edge != key:
                    incident_links.append(edge)
        if len(incident_links) == 1:
            end_c_link[key] = incident_links[0]
            end_cc_link[key] = incident_links[0]
        else:
            cwise = {}
            ccwise = {}
            origin = (vertices[value])
            for k in key:
                if k != value:
                    prev = vertices[k]
            #prev = (vertices[k for k in key if k != value])
            for incident_link in incident_links:
                if vertices[incident_link[0]] != origin:
                    inc = vertices[incident_link[0]]
                else:
                    inc = vertices[incident_link[1]]
                rotation = (prev[0]-origin[0])*(inc[1] - origin[1]) - (prev[1]-origin[1])*(inc[0]-origin[0]) #Cross Product
                if rotation >= 0:
                    #CCWise
                    ccwise[rotation] = incident_link
                else:
                    #CWise
                    cwise[rotation] = incident_link
            if len(cwise.keys()) != 0 and len(ccwise.keys()) !=0:
                end_c_link[key] = cwise[max(cwise.keys(), key=float)]
                end_cc_link[key] = ccwise[max(ccwise.keys(), key=float)]
            elif len(cwise.keys()) != 0:
                end_c_link[key] = cwise[min(cwise.keys(), key=float)]
                end_cc_link[key] = cwise[max(cwise.keys(), key=float)]
    
    #Pack the WED
    wed['node2link'] = node2link
    wed['region2link'] = region2link
    wed['start_node'] = startnode
    wed['end_node'] = end_node
    wed['right_region'] = right_region
    wed['left_region'] = left_region
    wed['start_c_link'] = start_c_link
    wed['start_cc_link'] = start_cc_link
    wed['end_c_link'] = end_c_link
    wed['end_cc_link'] = end_cc_link
    
    return wed
```
